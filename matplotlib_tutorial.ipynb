```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#adjusting figure size

plt.figure(figsize=(4,2),dpi=100)

#shorthand notation '[color][marker][line]' example "r.--"

plt.plot([1,2,3,5,7],[3,2,1,5,2],"r1--",label="f(x)=2x",markerfacecolor="black")

plt.title("just a random graph",fontdict={'fontname': "Comic Sans MS","fontsize":20})

plt.xlabel("x label")
plt.ylabel("y label")

#adjusting the scale of the y axis (values)
plt.yticks([0,1,2,3,4,5])

#adding the legend to the graph
plt.legend()

#saving the graphe in the current directory
plt.savefig("mygraph.png",dpi=400)

plt.show()
```

# Bar charts


```
labels=["A","B","C"]
values=[3,2,4]

plt.figure(figsize=(3,2),dpi=150)

bars=plt.bar(labels,values,label="x and y")

plt.xlabel("labels",fontdict={"fontsize":12})
plt.ylabel("values",fontdict={"fontsize":12})

plt.yticks([0,1,2,3,4])

plt.title("This is a bar graph",fontdict={"fontsize":20})

bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")

plt.legend()

plt.show()
```

#Working with FIFA Data

```
fifa=pd.read_csv("fifa_data.csv")

fifa.head()
```

```
bins=list(range(0,110,10))

plt.hist(fifa.Overall,bins=bins,color="#2d9596")

plt.xticks(bins)

plt.xlabel("Overall skill")
plt.ylabel("number of players")

plt.title("distribution of players skill in fifa 2017",fontdict = {'family': 'serif',
        'color':  'darkred',
        'weight': 'bold',
        'size': 16,
        })
plt.show()
```

# Pie Chart

```
lefties=fifa.loc[fifa['Preferred Foot']=="Left"].count()[0]
righties=fifa.loc[fifa['Preferred Foot']=="Right"].count()[0]

labels=["Left footers","Right footers"]
plt.pie([lefties,righties],colors=["#b5e7a0","#bdcebe"],labels=labels, autopct="%.2f %%")

plt.title("Foot preference of FIFA players")

plt.show()
```

```
weights=list(fifa.Weight)
weights=[x[:-3] for x in weights if type(x)!=float]
weights=[int(x)/2.05 for x in weights]


light_weight=[x for x in weights if x<65]
medium_weight=[x for x in weights if x<75]
heavy_weight=[x for x in weights if x>=75]


labels=["under 65","under 75","over 75"]

weights=[len(light_weight),len(medium_weight),len(heavy_weight)]

plt.style.use('ggplot')

explode=(0.2,0.2,0.1)
plt.pie(weights,labels=labels, autopct="%.2f %%", pctdistance=0.8, explode=explode)
plt.title("Weight Distribution of FIFA Players (in kg)")

plt.show()
```

# Box plot

```
plt.style.use("default")

barcelona=fifa.loc[fifa.Club=="FC Barcelona"]["Overall"]
madrid=fifa.loc[fifa.Club=="Real Madrid"]["Overall"]
chelsea=fifa.loc[fifa.Club=="Chelsea"]["Overall"]

boxes=plt.boxplot([barcelona,madrid,chelsea],labels=["barcelona","madrid","chelsea"],patch_artist=True)

for box in boxes['boxes']:
  box.set(color="#abcdef",linewidth=2)
  box.set(facecolor="#e9e0e0")
plt.title("FIFA Clubs overall players rating")

plt.ylabel("FIFA overall ratings")

plt.show()
```
